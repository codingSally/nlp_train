# 中文文件train.json处理

```python
def read_json(file_name):
    
    with open(file_name, 'rb') as f:
        json_string = json.load(f)
        text_lst = [context for data in json_string['data'] for paragraphs in data['paragraphs'] for context in paragraphs['context']]
        
        text = ''.join(text_lst)
        return text
```

```python
text = read_json('./train.json')

words = list(jieba.cut(text))

words[:20]
```
